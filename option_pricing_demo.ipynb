```python
import sys
import os
sys.path.append(os.getcwd())

import numpy as np
import matplotlib.pyplot as plt

from black_scholes import (
    call_price,
    put_price,
    call_delta,
    put_delta,
    gamma,
    vega,
    theta,
    rho,
    get_stock_price_and_vol,
)
```

```python
ticker = "AAPL"

S, sigma = get_stock_price_and_vol(ticker, period="1y")

r = 0.04
T = 30 / 365
K = S

S, sigma, r, T, K
```

```python
S_grid = np.linspace(0.5 * S, 1.5 * S, 80)
call_prices = [call_price(s, K, r, sigma, T) for s in S_grid]

plt.figure(figsize=(8, 5))
plt.plot(S_grid, call_prices, linewidth=2)
plt.axvline(S, linestyle="--", color="gray", label=f"Current S = {S:.2f}")
plt.xlabel("Stock Price S")
plt.ylabel("Call Price")
plt.title(f"Call Price vs Stock Price (AAPL, ATM, 30 days)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
S_grid = np.linspace(0.5 * S, 1.5 * S, 80)
call_deltas = [call_delta(s, K, r, sigma, T) for s in S_grid]

plt.figure(figsize=(8, 5))
plt.plot(S_grid, call_deltas, linewidth=2)
plt.axvline(S, linestyle="--", color="gray", label=f"Current S = {S:.2f}")
plt.xlabel("Stock Price S")
plt.ylabel("Call Delta")
plt.title(f"Call Delta vs Stock Price (AAPL, ATM, 30 days)")
plt.ylim(-0.05, 1.05)
plt.grid(True)
plt.legend()
plt.show()
```
